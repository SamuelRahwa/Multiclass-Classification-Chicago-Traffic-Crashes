```python
# import packages
import pandas as pd 
import numpy as np 

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('ggplot')

from matplotlib.pyplot import figure
matplotlib.rcParams['figure.figsize'] = (18,10)
matplotlib.rcParams.update({'font.size': 22})

import seaborn as sns 
sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(18, 10)})

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Final_Chicago_Traffic_Crashes_Cleaned.csv') 
display(data.shape)
data.info()
```

```python
data[['TRAFFIC_CONTROL_DEVICE', 'DEVICE_CONDITION',
       'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE',
       'TRAFFICWAY_TYPE', 'ALIGNMENT', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT',
       'REPORT_TYPE', 'CRASH_TYPE', 'DAMAGE', 
       'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE',
       'STREET_DIRECTION', 'STREET_NAME']] = data[['TRAFFIC_CONTROL_DEVICE', 'DEVICE_CONDITION',
       'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE',
       'TRAFFICWAY_TYPE', 'ALIGNMENT', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT',
       'REPORT_TYPE', 'CRASH_TYPE', 'DAMAGE', 
       'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE',
       'STREET_DIRECTION', 'STREET_NAME']].astype(str)
```

```python
plt.figure(figsize =(18,10))
plt.xticks()

ax = sns.countplot(x = data['CRASH_TYPE'], data = data, order = data['CRASH_TYPE'].value_counts().index, palette = 'muted')
plt.xlabel('Crash Type')
plt.ylabel('Type of Injury')
plt.title('Injuries vs No Injuries')
plt.show()
```

```python
ax = sns.countplot(x= "CRASH_HOUR", hue = 'CRASH_TYPE', data = data)
plt.xlabel('Time of Crash')
plt.ylabel('Number of Injuries vs No Injuries')
plt.title('Crash Type by the Hour')
plt.show()
```

```python
ax = sns.countplot(x= "NUM_UNITS", hue = 'CRASH_TYPE', data = data)
plt.xlabel('Num Unites')
plt.ylabel('Number of Injuries vs No Injuries')
plt.title('Crash Type by Num Units')
plt.show()
```
